```python
# creates a signature matrix for the given dataframe of mfcc values and the permutation matrix
def Signature_Matrix(df_new, d):
    res = np.unique(df_new.iloc[0, :])
    res
    hash=d.columns
    counter=99999
    s=[]
    sig=pd.DataFrame()
    
    for x in range(0,len(df_new.index)):
        res = np.unique(df_new.iloc[x, 1:201])#controlling the number of columns
        print(res)
        for i in hash:
            for j in res:
                # print(d.iloc[j, i])
                loc=d.iloc[j, i]
                if loc<counter:
                    counter=loc
            s.append(counter)
                
        counter=99999
        sig[x]=s
        s.clear()
    
    
    return sig
sig=Signature_Matrix(df_new, d)
sig   
```

```python
# returns a list of hash buckets/tables
def Hash_Bucket(sig, s=None):
  bucket=[{}, {}, {}, {}]
  x=0
  if s==None:
   while(x!=8):
    for j in bucket:
      for i in range(0,len(sig.columns)):
          num=sig[i][x]+sig[i][x+1]
          
          if num in j.keys():
            if isinstance(j[num], list):
              j[num].append(sound[i])
            else:
              temp_list = [j[num]]
              temp_list.append(sound[i])
              j[num] = temp_list
          else:
            j[num] = sound[i]
      x=x+2
  else:
    for j in bucket:
      for i in range(0,len(sig.columns)):
          num=sig[i][x]+sig[i][x+1]
          j[num] = s
      x=x+2        
 
  return bucket

# list of hash tables
bucket=Hash_Bucket(sig)


bucket1=bucket[0]
bucket2=bucket[1]
bucket3=bucket[2]
bucket4=bucket[3]
# printing one hash table/bucket
bucket4
```

```python
# to find the key(the sum of bands) of a value(audio) in a list of dictionaries
def find_key(val, buck=bucket):
    keys=[]
    for j in buck:  
      for key, value in j.items():
        if type(value)==list:
         for i in value:
          if i==val:
            keys.append(key)
        else:
          if value==val:
            keys.append(key)    
       
    return keys
    return "key doesn't exist"
  
print(find_key('001039.mp3', bucket))
```

```python
# to calculate jaccard similarity between two lists contaiming keys as bands sum
def jaccard_similarity(list1, list2):
    intersection = len(list(set(list1).intersection(list2)))
    union = (len(list1) + len(list2)) - intersection
    return float(intersection / union)*100

# test jacard similarity
l1=find_key(sound[0])
l2=find_key(sound[1]) 
jaccard_similarity(l1, l2)   
```

```python
# to find similarity between two sounds only
def similarity(sound1,sound2):
  l1=find_key(sound1)
  l2=find_key(sound2) 
  return jaccard_similarity(l1, l2)

similarity(sound[0],sound[1])  
```
